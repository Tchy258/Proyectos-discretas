###Falta poner los tiempos promedios de ejecución en los arrays

```python
import matplotlib.pyplot as plt
import numpy as np

# Para una cantidad de 14:3 
cantidadProposiciones = np.arange(3,16,3)

# Tiempo promedio de ejecucion de cada SAT solver
tiempoSAT = np.array([0.0011088318499929528,0.010302556249996542,0.08956253620000894,1.017059680899996,9.9648693438])
tiempoMinisat = np.array([6.360324999548084e-05,7.488784999054587e-05,0.0001020860500091203,0.00012582000001089,0.0001499237500070194])

# Gráfico
fig, ax = plt.subplots()
# Leyendas
ax.plot(cantidadProposiciones, tiempoSAT, label = 'SAT1')
ax.plot(cantidadProposiciones, tiempoMinisat, label = 'Minisat22')
ax.legend(loc = 'upper right')
# Ejes
plt.xlabel('Cantidad de proposiciones')
plt.ylabel('Tiempo de ejecución')
# Titulo
plt.title('Gráfico tiempo de ejecución vs cantidad de proposiciones')
plt.show()
```
