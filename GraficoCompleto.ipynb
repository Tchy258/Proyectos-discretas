```python
import matplotlib.pyplot as plt
import numpy as np

# Para (clausulas:proposiciones) -> 14:3, 28:6, 42:9, 56:12, 70:15 y 84:18
cantidadProposiciones = np.arange(3,19,3)

# Tiempo promedio de ejecucion de cada SAT solver
tiempoSAT = np.array([0.0011088318499929528,0.010302556249996542,0.08956253620000894,1.017059680899996,9.9648693438,96.98589284375001])
tiempoMinisat = np.array([6.360324999548084e-05,7.488784999054587e-05,0.0001020860500091203,0.00012582000001089,0.0001499237500070194,0.00020218290004549999])

# Gr치fico
plt.rcParams["figure.figsize"] = (10, 5)
plt.rc('font', size=15)  
fig, ax = plt.subplots()
# Leyendas
ax.plot(cantidadProposiciones, tiempoSAT, label = 'SAT solver propio')
ax.plot(cantidadProposiciones, tiempoMinisat, label = 'SAT solver Minisat22')
ax.legend(loc = 'upper left')
# Ejes
plt.xlabel('Cantidad de proposiciones')
plt.ylabel('Tiempo de ejecuci칩n promedio [s]')
# Titulo
plt.title('Gr치fico tiempo de ejecuci칩n vs cantidad de proposiciones')
plt.show()
```
